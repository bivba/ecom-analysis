```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import polars as pl
```

```python
nan_brand = data_nov.filter(pl.col('category_code').is_null()).select('brand')
nan_brand.collect()
```

```python
category_brand = data_nov.filter(pl.col('brand').is_in(nan_brand.select('brand').collect().to_series())).select(['category_code', 'brand'])
category_brand.collect()
```

```python
category_by_brand = category_brand.group_by('brand').agg(
    pl.col('category_code').mode().first().alias('common_category')
)
brand_by_category = category_brand.group_by('category_code').agg(
    pl.col('brand').mode().first().alias('common_brand')
)
print(category_by_brand.collect_schema())
print(brand_by_category.collect_schema())
print(data_nov.collect_schema())
data_nov = (
    data_nov.join(category_by_brand, on='brand', how='left').with_columns(pl.coalesce('category_code', 'common_category').alias('category_code'))
)
data_nov = (
    data_nov.join(brand_by_category, on='category_code', how='left').with_columns(pl.coalesce('brand', 'common_brand').alias('brand'))
)
data_nov = data_nov.drop("common_category", "common_brand")
data_nov.head(10).collect()
```

```python
product_to_category = data_nov.group_by('product_id').agg(pl.col('category_id').mode().first().alias('common_category'))

product_to_brand = data_nov.group_by('product_id').agg(pl.col('brand').mode().first().alias('common_brand'))

data_nov = (
    data_nov.join(product_to_category, on='product_id', how='left').with_columns(pl.coalesce('product_id', 'common_category').alias('product_id'))
)
data_nov = (
    data_nov.join(product_to_brand, on='product_id', how='left').with_columns(pl.coalesce('product_id', 'common_brand').alias('product_id'))
)
data_nov = data_nov.drop("common_category", "common_brand")
data_nov.head(10).collect()
```

```python
category_id_code = (
    data_nov.select(['category_id', 'category_code'])
    .filter(pl.col('category_code').is_not_null())
    .unique()
)
data_nov = data_nov.join(
    category_id_code, 
    on='category_id', 
    how='left',
    suffix='_filled'
).with_columns(
    pl.coalesce(['category_code', 'category_code_filled']).alias('category_code')
)
data_nov = data_nov.drop('category_code_filled')
data_nov.head(10).collect()
```

```python
null_counts = data_nov.null_count().collect()

for col in null_counts.columns:
    print(f"{col}: {null_counts[col][0]} пропусков")
```

```python
from sklearn.impute import KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
def impute(df: pl.LazyFrame, target_columns=['category_code', 'brand'], n_neighbours=3):
    tmp = df.collect()
    event_time = tmp.get_column('event_time')
    df_imp = tmp

    # df_imp['log_price'] = np.log1p(df_imp['price'])
    # df_imp['event_time'] = pd.to_datetime(df_imp['event_time'])
    # df_imp['day'] = df_imp['event_time'].dt.day
    # df_imp['month'] = df_imp['event_time'].dt.month
    # df_imp['year'] = df_imp['event_time'].dt.year
    # df_imp.drop('event_time', axis=1, inplace=True)

    df_imp = df_imp.with_columns(
        pl.col('event_time').str.strip_suffix(" UTC").str.to_datetime().alias('event_time_dt')
    ).with_columns([
        pl.col('price').log1p().alias('log_price'),
        pl.col('event_time_dt').dt.day().alias('day'),
        pl.col('event_time_dt').dt.month().alias('month'),
        pl.col('event_time_dt').dt.year().alias('year'),
        pl.col('event_time_dt').dt.weekday().alias('weekday'),
        pl.col('event_time_dt').dt.ordinal_day().alias('day_of_year')
    ]).drop(['event_time', 'event_time_dt'])
    
    cat_cols = ['event_type', 'user_session'] + target_columns
    encoders = {}

    encoding_expr = []

    df_imp = df_imp.with_columns(
        [pl.col(c).fill_null('MISSING') for c in cat_cols]
    )

    for col in cat_cols:
        oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
        column = df_imp.select(col).to_numpy().reshape(-1, 1)
        encoded = oe.fit_transform(column).flatten()
        encoders[col] = oe
        df_imp = df_imp.with_columns(pl.Series(name=col, values=encoded))
    
    imputer = KNNImputer(n_neighbors=n_neighbours, weights='distance', metric='nan_euclidean', missing_values=-1)
    imputed_data = imputer.fit_transform(df_imp.to_numpy())

    #df_original = df.copy()

    df_imputed_pl = pl.DataFrame(imputed_data, schema=df_imp.columns, orient='row')
    decode_expr = []

    for col in cat_cols:
        oe = encoders[col]
        categories = oe.categories_[0]
        max_valid = len(categories) - 1

        mapping = {str(i): val for i, val in enumerate(categories)}
        expr = (
            pl.col(col)
            .round(0)
            .cast(pl.Int64)
            .cast(pl.String)
            .replace(mapping)
            .alias(col)
        )
        decode_expr.append(expr)
    
    df_final = df_imputed_pl.with_columns(decode_expr)

    df_final = df_final.with_columns(event_time)

    return df_final.lazy()

    # for i, col in enumerate(df_imp.columns):
    #     imputed_series = pl.Series(imputed_data[:, i])
        
    #     df_original = df_original.with_columns(
    #         pl.lit(imputed_series).alias(col)
    #     )
    
    # for col in cat_cols:
    #     df_original[col] = df_original[col].round().astype(int)

    #     max_valid = len(encoders[col].categories_[0]) - 1
    #     df_original[col] = df_original[col].clip(0, max_valid)
    #     df_original[col] = encoders[col].inverse_transform(df_original[[col]])[:, 0]
    
    
    #return df_original
```

```python
df_imputed = impute(data_nov)
df_imputed.head(10).collect()
```

```python
# session_df = df_imputed.groupby('user_session').agg(
#     event_number=('event_type', 'count'),
#     session_duration=('event_time', lambda x: max(x) - min(x)),
#     carted=('event_type', lambda x: 1 if 'cart' in x.values else 0),
#     purchased=('event_type', lambda x: 1 if 'purchase' in x.values else 0),
#     most_searched_catalog=('catalog', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     most_searched_brand=('brand', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     mean_price_searched=('price', 'mean')
# )
# session_df

session_df = df_imputed.group_by(pl.col('user_session')).agg(
    event_number=pl.col('event_type').count(),
    session_duration=(pl.col('event_time').max() - pl.col('event_time').min()).dt.total_seconds().cast(pl.Int64),
    carted=(pl.col('event_type') == 'cart').any().cast(pl.Int8),
    purchased=(pl.col('event_type') == 'purchase').any().cast(pl.Int8),
    most_searched_catalog=pl.col('catalog').mode().first(),
    most_searched_brand=pl.col('brand').mode().first(),
    mean_price_searched=pl.col('price').mean()
)
session_df.head(10).collect()
```

```python
# group_df = df_imputed.groupby('product_id').agg(
#     category_id=('category_id', 'max'),
#     category_code=('category_code', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     brand=('brand', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     mean_price=('price', 'mean'),
#     popularity=('user_session', lambda x: x.nunique()),
#     times_purchased=('event_type', lambda x: (x == 'purchase').sum())
# )
# group_df['popularity'] = (group_df['popularity'] - group_df['popularity'].min()) / (group_df['popularity'].max() - group_df['popularity'].min())
# group_df

group_df = df_imputed.group_by('product_id').agg(
    category_id=pl.col('category_id').max(),
    category_code=pl.col('category_code').mode().first(),
    brand=pl.col('brand').mode().first(),
    mean_price=pl.col('price').mean(),
    popularity=pl.col('user_session').n_unique(),
    times_purchased=(pl.col('event_type') == 'purchase').sum(),
    times_carted=(pl.col('event_type') == 'cart').sum()
)
group_df = group_df.with_columns((pl.col('popularity') - pl.col('popularity').min()) / (pl.col('popularity').max() - pl.col('popularity').min()))
group_df.head(10).collect()
```

```python
# users_df = df_imputed.groupby('user_id').agg(
#     number_of_sessions=('user_session', 'count'),
#     most_viewed_category_id=('category_id', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     most_viewed_category_code=('category_code', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     most_viewed_brand=('brand', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
#     times_purchased=('event_type', lambda x: (x == 'purchase').sum()),
#     times_cart=('event_time', lambda x: (x == 'cart').sum())
# )
# spend = df_imputed[df_imputed['event_type'] == 'purchase'].groupby('user_id')['price'].sum()
# users_df = pd.merge(users_df, spend, on='user_id', how='left').fillna(0)
# users_df

users_df = df_imputed.group_by('user_id').agg(
    number_of_sessions=pl.col('user_session').count(),
    most_viewed_category_code=pl.col('category_code').mode().first(),
    most_viewed_brand=pl.col('brand').mode().first(),
    times_purchased=(pl.col('event_type') == 'purchase').any().cast(pl.Int8),
    times_carted=(pl.col('event_type') == 'cart').any().cast(pl.Int8)
)
spend = df_imputed.filter(pl.col('event_type') == 'purchase').group_by('user_id').agg(pl.col('price').sum().alias('total_spend'))
users_df = users_df.join(spend, on='user_id', how='left').fill_null(0)
users_df.head(10).collect()
```
